```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv(r'C:\Users\jdffc\Downloads\DataProjects\AirbnbToronto\data\airbnb_cleaned.csv')
```

### Average Price by Neighborhood

```python
# Visualizing the Top 10 Most Expensive Neighborhoods in Toronto

# Set the figure size for the plot (width = 12, height = 6)
# This makes the chart bigger and easier to read.

plt.figure(figsize=(12,6))

# Group the dataset by neighbourhood and calculate the average price in each one
# Then, sort the average prices in descending order and take the top 10.

neigh_price=df.groupby('neighbourhood_cleansed') ['price'].mean().sort_values(ascending=False).head(10)

# Create a horizontal bar chart using seaborn
# x = average price values, y = neighborhood names, with a 'coolwarm' color palette

sns.barplot(x=neigh_price.values, y=neigh_price.index, palette='coolwarm')

# Add a descriptive title and axis labels for clarity

plt.title('Top 10 Most Expensive Neighbourhoods in Toronto')
plt.xlabel('Average Price in CAD')
plt.ylabel('Neighbourhood')

# Display the final plot

plt.savefig(r'C:\Users\jdffc\Downloads\DataProjects\AirbnbToronto\images\top_neighbourhoods_toronto.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Price Distribution

```python
# Visualizing the Distribution of Airbnb Prices in Toronto

# Set the overall figure size (width = 8, height = 5)
# This controls how large the chart appears.

plt.figure(figsize=(8,5))

# Create a histogram of Airbnb prices using seaborn's histplot
# 'bins=50' divides the price range into 50 equal intervals (bars)
# 'kde=True' adds a smooth curve (Kernel Density Estimate) to show the distribution shape

sns.histplot(df['price'], bins=50, kde=True)

# Add a descriptive title to explain what the chart shows

plt.title('Distribution of Airbnb Prices in Toronto')

# Label the x-axis to show that it represents price values in CAD

plt.xlabel('Prices in CAD')

# Display the final plot

plt.savefig(r'C:\Users\jdffc\Downloads\DataProjects\AirbnbToronto\images\distribution_prices_toronto.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Room Type vs Price

```python
# Visualizing the Relationship Between Room Type and Price

# Set the figure size (width = 7, height = 4)
# This adjusts how big the plot appears in the output.

plt.figure(figsize=(7,4))

# Create a boxplot using seaborn to compare prices across different room types
# 'x' = categories (room types), 'y' = numeric values (prices)
# Each box shows the median, quartiles, and potential outliers for that category.

sns.boxplot(data=df, x='room_type', y='price')

# Add a descriptive title so viewers know what the chart represents

plt.title('Room Type vs Price')

#Display the boxplot

plt.savefig(r'C:\Users\jdffc\Downloads\DataProjects\AirbnbToronto\images\Room_type_price.png', dpi=300, bbox_inches='tight')
plt.show()
```

#### Correlation Heatmap

```python
# Visualizing Correlation Between Numerical Features

# Set the figure size (width = 8, height = 5)
# This defines how big the heatmap will appear.

plt.figure(figsize=(8,5))

# Create a correlation heatmap using seaborn
# 'df.corr(numeric_only=True)' calculates correlation values only for numeric columns
# 'annot=True' displays the actual correlation numbers inside the boxes
# 'cmap="Blues"' gives the heatmap a blue color gradient for better readability

sns.heatmap(df.corr(numeric_only=True), cmap='Blues')

# Add a clear, descriptive title for the visualization

plt.title('Correlation Heatmap')

#Display the heatmap

plt.savefig(r'C:\Users\jdffc\Downloads/DataProjects/AirbnbToronto/images\correlation.png', dpi=300, bbox_inches='tight')
plt.show()
```

### Key Insights
- Downtown neighborhoods tend to have the highest average prices.
- Entire homes/apartments cost more than private rooms.
- Most listings are priced below 300 CAD per night.
- Higher ratings slightly correlate with higher prices.
- Review scores and availability have weak correlation with price.
